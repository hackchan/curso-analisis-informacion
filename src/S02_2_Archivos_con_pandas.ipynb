# Lectura de Archivos con Pandas

Este notebook te guiará a través de los conceptos fundamentales y avanzados para leer archivos CSV y Excel utilizando la poderosa librería `pandas` en Python. Cubriremos:

1.  **Lectura Básica de CSV y Excel**
2.  **Manejo de Estructuras Complejas** (delimitadores, encabezados, saltos de línea, hojas de Excel específicas)
3.  **Gestión de Codificaciones (Encodings)**

```python
import pandas as pd
import numpy as np
import os
```

## 1. Lectura Básica de Archivos CSV

La función `pd.read_csv()` es la herramienta principal para leer archivos CSV. Su uso más básico es simplemente pasar la ruta del archivo.

```python
# Crear un archivo CSV simple para demostración
data_csv_simple = {
    'Nombre': ['Alice', 'Bob', 'Charlie'],
    'Edad': [24, 27, 22],
    'Ciudad': ['Nueva York', 'Londres', 'París']
}
df_simple = pd.DataFrame(data_csv_simple)
df_simple.to_csv('datos_simples.csv', index=False)

print('Archivo "datos_simples.csv" creado.')
```

```python
# Leer el archivo CSV simple
df_leido_csv = pd.read_csv('datos_simples.csv')
print("Contenido del archivo CSV simple:")
display(df_leido_csv)
```

## 2. Lectura Básica de Archivos Excel

La función `pd.read_excel()` se utiliza para leer archivos Excel (.xlsx, .xls). Al igual que `read_csv`, su uso básico es pasar la ruta del archivo.

```python
# Crear un archivo Excel simple para demostración
data_excel_simple = {
    'Producto': ['Laptop', 'Mouse', 'Teclado'],
    'Precio': [1200, 25, 75],
    'Cantidad': [10, 50, 30]
}
df_excel_simple = pd.DataFrame(data_excel_simple)
df_excel_simple.to_excel('productos_simples.xlsx', index=False)

print('Archivo "productos_simples.xlsx" creado.')
```

```python
# Leer el archivo Excel simple
df_leido_excel = pd.read_excel('productos_simples.xlsx')
print("Contenido del archivo Excel simple:")
display(df_leido_excel)
```

## 3. Manejo de Estructuras Complejas en CSV

Los archivos CSV no siempre son tan limpios. `pd.read_csv()` ofrece varios parámetros para manejar estas situaciones:

*   `sep`: Especifica el delimitador (ej. `;`, `\t` para tabulaciones).
*   `header`: Fila a usar como nombres de columna (0-indexado), o `None` si no hay encabezado.
*   `skiprows`: Número de filas a ignorar al principio del archivo.
*   `names`: Lista de nombres de columna para usar si no hay encabezado o quieres renombrarlos.
*   `decimal`: Carácter a reconocer como separador decimal (ej. `,` para español).
*   `thousands`: Carácter a reconocer como separador de miles.

```python
# Crear un archivo CSV con estructura compleja
contenido_csv_complejo = (
    "# Esto es un comentario\n"
    "Fecha;Ventas;Region\n"
    "2023-01-01;123,45;Norte\n"
    "2023-01-02;67,89;Sur\n"
    "2023-01-03;90,12;Este\n"
    "Final del reporte"
)
with open('ventas_complejas.csv', 'w', encoding='utf-8') as f:
    f.write(contenido_csv_complejo)

print('Archivo "ventas_complejas.csv" creado con estructura compleja.')
```

```python
# Leer el archivo CSV complejo:
# - `sep=';'` porque usa punto y coma como delimitador
# - `skiprows=[0, 5]` para ignorar la primera fila (comentario) y la última (final del reporte)
# - `decimal=','` para reconocer la coma como separador decimal

df_leido_complejo_csv = pd.read_csv(
    'ventas_complejas.csv',
    sep=';',
    skiprows=[0, 5],
    decimal=','
)
print("Contenido del archivo CSV complejo:")
display(df_leido_complejo_csv)
print("Tipos de datos después de la lectura:")
df_leido_complejo_csv.info()
```

## 4. Manejo de Estructuras Complejas en Excel

Los archivos Excel pueden tener múltiples hojas, o datos que no comienzan en la primera fila/columna. `pd.read_excel()` tiene parámetros como:

*   `sheet_name`: Nombre o índice de la hoja a leer. Puede ser una lista para leer varias hojas.
*   `header`: Similar a CSV, especifica la fila del encabezado.
*   `skiprows`: Similar a CSV.
*   `usecols`: Columnas específicas a importar (por índice o nombre).
*   `nrows`: Número de filas a leer.

```python
# Crear un archivo Excel con múltiples hojas y datos desalineados
with pd.ExcelWriter('reporte_multih.xlsx') as writer:
    # Hoja 1: 'Resumen Mensual'
    df_resumen = pd.DataFrame({
        'Mes': ['Enero', 'Febrero', 'Marzo'],
        'Ingresos': [10000, 12000, 11500]
    })
    df_resumen.to_excel(writer, sheet_name='Resumen Mensual', index=False)

    # Hoja 2: 'Detalle Ventas' con algunas filas de metadata arriba
    df_detalle = pd.DataFrame({
        'ID Transacción': [1, 2, 3],
        'Producto': ['A', 'B', 'C'],
        'Cantidad': [5, 2, 8]
    })
    # Simular metadata antes de los datos reales
    metadata_rows = pd.DataFrame([['Reporte de Ventas', np.nan, np.nan], ['Generado el:', '2023-01-01', np.nan], [np.nan, np.nan, np.nan]])
    df_final_detalle = pd.concat([metadata_rows, df_detalle], ignore_index=True)
    df_final_detalle.to_excel(writer, sheet_name='Detalle Ventas', index=False, header=False) # Escribimos sin header para manejarlo con skiprows

print('Archivo "reporte_multih.xlsx" creado.')
```

```python
# Leer una hoja específica por nombre
df_resumen_mensual = pd.read_excel('reporte_multih.xlsx', sheet_name='Resumen Mensual')
print("Datos de la hoja 'Resumen Mensual':")
display(df_resumen_mensual)
```

```python
# Leer la hoja 'Detalle Ventas' omitiendo las primeras filas y especificando el encabezado
df_detalle_ventas = pd.read_excel(
    'reporte_multih.xlsx',
    sheet_name='Detalle Ventas',
    skiprows=3, # Ignorar las primeras 3 filas de metadata
    header=None, # Indicamos que no hay header en la primera fila leída (ya que la 4ta fila es el header)
    names=['ID Transacción', 'Producto', 'Cantidad'] # Asignamos los nombres manualmente
)
print("Datos de la hoja 'Detalle Ventas' (con skiprows y nombres de columna):")
display(df_detalle_ventas)
```

## 5. Gestión de Codificaciones (Encodings)

La codificación de caracteres es crucial cuando se trabaja con texto. Si un archivo se guarda con una codificación (ej. `latin1`) y se lee con otra (ej. `utf-8`), los caracteres especiales (ñ, tildes, €, etc.) pueden aparecer corruptos o causar errores.

El parámetro `encoding` en `pd.read_csv()` (y otras funciones de lectura) es vital.

Algunas codificaciones comunes:
*   `utf-8`: La codificación más universal y recomendada.
*   `latin1` (ISO-8859-1): Común en sistemas europeos occidentales.
*   `cp1252`: Codificación de Windows para Europa Occidental (similar a `latin1`).
*   `utf-8-sig`: UTF-8 con una marca de orden de bytes (BOM), a veces usada por Excel para exportar.

```python
# Crear un archivo CSV con una codificación diferente (latin1) y caracteres especiales
contenido_latin1 = (
    "ID;Descripción;Valor\n"
    "1;Pequeño negocio;100\n"
    "2;Grandes áreas;250\n"
    "3;Fenómenos atmosféricos;50\n"
)
with open('datos_latin1.csv', 'w', encoding='latin1') as f:
    f.write(contenido_latin1)

print('Archivo "datos_latin1.csv" creado con codificación latin1.')
```

```python
# INTENTO FALLIDO: Leer el archivo sin especificar la codificación (pandas intenta utf-8 por defecto)
print("Intento de lectura sin encoding (esperando error o caracteres incorrectos):")
try:
    df_error_encoding = pd.read_csv('datos_latin1.csv', sep=';')
    display(df_error_encoding) # Si no da error, los caracteres estarán mal
except UnicodeDecodeError as e:
    print(f"¡Error de decodificación esperado! {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
# LECTURA CORRECTA: Especificar la codificación correcta
df_latin1_correcto = pd.read_csv('datos_latin1.csv', sep=';', encoding='latin1')
print("Lectura correcta con encoding='latin1':")
display(df_latin1_correcto)
```
