# SQL - Lenguaje Estructurado de Consulta (Structured Query Language)

SQL (Structured Query Language) es un lenguaje de programación estándar diseñado para gestionar y manipular bases de datos relacionales. Es el lenguaje más común utilizado para interactuar con sistemas de gestión de bases de datos (DBMS) como MySQL, PostgreSQL, Oracle, SQL Server y SQLite.

**¿Por qué es importante SQL?**

1.  **Dominio Global**: Es el estándar de facto para la gestión de bases de datos relacionales, lo que significa que las habilidades en SQL son altamente transferibles y demandadas en casi cualquier industria que maneje datos.
2.  **Gestión de Datos**: Permite a los usuarios almacenar, recuperar, modificar y eliminar datos de manera eficiente. Es fundamental para organizar grandes volúmenes de información.
3.  **Análisis y Reporte**: Facilita la extracción de datos específicos para análisis, la generación de informes y la toma de decisiones empresariales. Los analistas de datos, científicos de datos y desarrolladores lo utilizan para obtener información valiosa.
4.  **Integridad de Datos**: Ayuda a mantener la consistencia y precisión de los datos a través de reglas y restricciones, asegurando que la información sea fiable.
5.  **Comunicación con la Base de Datos**: Es la interfaz universal para comunicarse con las bases de datos, independientemente del sistema específico que se esté utilizando.

## Bases de Datos Relacionales y NoSQL

Para entender mejor SQL, es útil conocer los tipos de bases de datos que existen.

### Bases de Datos Relacionales

Las bases de datos relacionales son el modelo más común y están basadas en el modelo relacional propuesto por E.F. Codd. Se caracterizan por:

*   **Estructura Tabular**: Los datos se organizan en tablas (relaciones) compuestas por filas (registros) y columnas (atributos).
*   **Esquema Fijo**: Requieren un esquema predefinido antes de almacenar los datos. Cada columna tiene un tipo de dato específico.
*   **Claves**: Utilizan claves primarias y foráneas para establecer relaciones entre tablas, garantizando la integridad referencial de los datos.
*   **Consistencia (ACID)**: Priorizan la atomicidad, consistencia, aislamiento y durabilidad (ACID) de las transacciones, lo que las hace ideales para aplicaciones que requieren alta fiabilidad, como sistemas bancarios.
*   **SQL**: SQL es el lenguaje estándar para interactuar con bases de datos relacionales.

**Ejemplos**: MySQL, PostgreSQL, Oracle, SQL Server, **SQLite**.

### SQLite

**SQLite** es una base de datos relacional única por varias razones:

*   **Embebida**: No es un sistema cliente-servidor, sino que la base de datos completa reside en un único archivo en disco. Esto significa que no hay un proceso de servidor separado y la interacción se realiza directamente con el archivo.
*   **Autocontenida**: Requiere una configuración mínima o nula.
*   **Sin Servidor**: No necesita un servidor de base de datos en ejecución.
*   **Transaccional**: A pesar de su simplicidad, soporta transacciones ACID.
*   **Ligera**: Es muy compacta y eficiente, ideal para dispositivos móviles, aplicaciones de escritorio pequeñas, y como base de datos local para desarrollo o pruebas (como en este notebook).

### Bases de Datos NoSQL (Not Only SQL)

Las bases de datos NoSQL surgieron como una alternativa a las bases de datos relacionales para abordar nuevos desafíos, especialmente en el manejo de grandes volúmenes de datos no estructurados o semi-estructurados, y requisitos de escalabilidad horizontal. Se caracterizan por:

*   **Esquema Flexible**: No requieren un esquema fijo, lo que permite almacenar datos de diversas estructuras de forma dinámica.
*   **Diversidad de Modelos**: Incluyen varios tipos de modelos de datos:
    *   **Clave-Valor**: Como Redis, DynamoDB.
    *   **Documentales**: Almacenan datos en documentos JSON o XML, como MongoDB, CouchDB.
    *   **Orientadas a Columnas**: Como Cassandra, HBase.
    *   **Grafo**: Almacenan datos como nodos y relaciones, como Neo4j.
*   **Escalabilidad Horizontal**: Diseñadas para escalar fácilmente distribuyendo los datos en múltiples servidores (sharding).
*   **Disponibilidad y Tolerancia a Particiones (BASE)**: A menudo priorizan la disponibilidad y la tolerancia a particiones sobre la consistencia fuerte (modelo BASE: Basically Available, Soft state, Eventually consistent).

**¿Cuándo usar cada una?**

*   **Relacionales (SQL)**: Ideales cuando los datos son estructurados, las relaciones son claras y la integridad transaccional es crítica (ej. finanzas, inventarios, sistemas de gestión de clientes).
*   **NoSQL**: Ideales para datos no estructurados, aplicaciones que requieren alta escalabilidad y disponibilidad, o cuando el modelo de datos no se adapta bien a un esquema tabular (ej. redes sociales, IoT, análisis de big data, CMS).

## Tipos de Consultas SQL:

*   **DDL (Data Definition Language)**: Para definir y modificar la estructura de la base de datos.
*   **DML (Data Manipulation Language)**: Para insertar, actualizar y eliminar datos.
*   **DQL (Data Query Language)**: Para consultar y recuperar datos.
*   **DCL (Data Control Language)**: Para gestionar permisos de acceso (explicación conceptual, ya que SQLite tiene limitaciones).
*   **TCL (Transaction Control Language)**: Para gestionar transacciones.


## 1. Configuración de la Base de Datos SQLite

````python
import sqlite3
import pandas as pd # Utilizaremos pandas para una visualización más amigable de los resultados

# Conectar a una base de datos en memoria (se destruye al finalizar la sesión)
conn = sqlite3.connect(':memory:')
# Conectar a una base de datos basada en archivo (se guardará en 'auditoria.db')
#conn = sqlite3.connect('auditoria.db')
c = conn.cursor()

print("Base de datos SQLite en memoria creada exitosamente.")
````

## 2. DDL (Data Definition Language): Definiendo la Estructura

El DDL se utiliza para definir y modificar la estructura de los objetos de la base de datos, como tablas, índices y vistas. Las operaciones DDL incluyen `CREATE`, `ALTER` y `DROP`.

````python
# Crear la tabla 'auditorias'
c.execute('''
    CREATE TABLE auditorias (
        id_auditoria INTEGER PRIMARY KEY AUTOINCREMENT,
        nombre_auditoria TEXT NOT NULL,
        fecha_inicio DATE NOT NULL,
        fecha_fin DATE,
        estado TEXT NOT NULL
    );
''')

# Crear la tabla 'funcionarios'
c.execute('''
    CREATE TABLE funcionarios (
        id_funcionario INTEGER PRIMARY KEY AUTOINCREMENT,
        nombre TEXT NOT NULL,
        cargo TEXT NOT NULL,
        departamento TEXT NOT NULL
    );
''')

# Crear la tabla 'hallazgos'
c.execute('''
    CREATE TABLE hallazgos (
        id_hallazgo INTEGER PRIMARY KEY AUTOINCREMENT,
        id_auditoria INTEGER NOT NULL,
        descripcion TEXT NOT NULL,
        severidad TEXT NOT NULL,
        monto_afectado REAL,
        fecha_identificacion DATE NOT NULL,
        FOREIGN KEY (id_auditoria) REFERENCES auditorias(id_auditoria)
    );
''')

print("Tablas 'auditorias', 'funcionarios' y 'hallazgos' creadas exitosamente.")
````

### 2.2. `ALTER TABLE`: Modificando la Estructura de una Tabla

El comando `ALTER TABLE` permite añadir, modificar o eliminar columnas de una tabla existente. En SQLite, las operaciones de `ALTER TABLE` son algo limitadas en comparación con otros sistemas de gestión de bases de datos, principalmente se usa para añadir columnas o renombrar tablas.

**Ejemplo: Añadir una columna a la tabla `funcionarios`**

````python
c.execute("ALTER TABLE funcionarios ADD COLUMN email TEXT;")
print("Columna 'email' añadida a la tabla 'funcionarios'.")

# Para verificar la nueva estructura (usando una consulta de esquema)
# En SQLite no hay un DESCRIBE fácil, pero podemos consultar la tabla 'sqlite_master'
result = c.execute("PRAGMA table_info(funcionarios);").fetchall()
print("\nNueva estructura de la tabla 'funcionarios':")
for row in result:
    print(row)
````

### 2.3. `DROP TABLE`: Eliminando Tablas

El comando `DROP TABLE` elimina una tabla existente y todos sus datos de la base de datos de forma permanente.

**Ejemplo: Crear una tabla temporal y luego eliminarla**

````python
c.execute('''
    CREATE TABLE tabla_temporal (
        id INTEGER PRIMARY KEY,
        dato TEXT
    );
''')
print("Tabla 'tabla_temporal' creada.")

c.execute("DROP TABLE tabla_temporal;")
print("Tabla 'tabla_temporal' eliminada.")
````

## 3. DML (Data Manipulation Language): Manipulando Datos

El DML se utiliza para manipular los datos dentro de las tablas de la base de datos. Los comandos DML más comunes son `INSERT`, `UPDATE` y `DELETE`.

### 3.1. `INSERT INTO`: Insertando Datos

El comando `INSERT INTO` se utiliza para añadir nuevas filas (registros) a una tabla. Puedes especificar las columnas en las que insertar o, si vas a insertar valores para todas las columnas en el orden correcto, puedes omitir la lista de columnas.

**Ejemplo: Insertar datos en la tabla `auditorias`**

````python
c.execute("""
    INSERT INTO auditorias (nombre_auditoria, fecha_inicio, fecha_fin, estado)
    VALUES
        ('Auditoría Financiera 2023 MinHacienda', '2023-01-15', '2023-06-30', 'Finalizada'),
        ('Auditoría de Cumplimiento Contratación Pública IDU', '2023-03-01', '2023-09-15', 'En Proceso'),
        ('Auditoría de Desempeño Gestión Ambiental CAR', '2023-05-10', NULL, 'Planeada'),
        ('Auditoría al Programa de Alimentación Escolar (PAE)', '2022-11-01', '2023-04-30', 'Finalizada');
""")
conn.commit() # Confirmar los cambios en la base de datos
print("Datos insertados en la tabla 'auditorias' exitosamente.")
````

**Ejemplo: Insertar datos en la tabla `funcionarios`**

````python
c.execute("""
    INSERT INTO funcionarios (nombre, cargo, departamento, email)
    VALUES
        ('Ana María Gómez', 'Auditor Senior', 'Financiero', 'ana.gomez@contraloria.gov.co'),
        ('Carlos Pérez Rojas', 'Jefe de Equipo de Auditoría', 'Contratación', 'carlos.perez@contraloria.gov.co'),
        ('Laura Botero Vélez', 'Auditor Junior', 'Desempeño', 'laura.botero@contraloria.gov.co'),
        ('Miguel Ángel López', 'Coordinador de Proyectos', 'Financiero', 'miguel.lopez@contraloria.gov.co');
""")
conn.commit()
print("Datos insertados en la tabla 'funcionarios' exitosamente.")
````

**Ejemplo: Insertar datos en la tabla `hallazgos`**

````python
c.execute("""
    INSERT INTO hallazgos (id_auditoria, descripcion, severidad, monto_afectado, fecha_identificacion)
    VALUES
        (1, 'Deficiencias en la conciliación de cuentas bancarias', 'Alta', 150000000.00, '2023-04-20'),
        (1, 'Incumplimiento de procedimientos internos para pagos', 'Media', 50000000.00, '2023-05-10'),
        (2, 'Adjudicación de contrato sin cumplir requisitos de publicidad', 'Alta', 5000000000.00, '2023-08-01'),
        (4, 'Alimentos entregados no cumplen con especificaciones técnicas', 'Grave', 750000000.00, '2023-03-15');
""")
conn.commit()
print("Datos insertados en la tabla 'hallazgos' exitosamente.")
````

### 3.2. `UPDATE`: Modificando Datos Existentes

El comando `UPDATE` se utiliza para modificar los valores de columnas en filas existentes de una tabla. Es crucial usar la cláusula `WHERE` para especificar qué filas deben ser actualizadas; de lo contrario, ¡se actualizarían todas las filas de la tabla!

**Ejemplo: Actualizar el estado de una auditoría y el monto afectado de un hallazgo**

````python
# Cambiar el estado de la auditoría 'Auditoría de Desempeño Gestión Ambiental CAR' a 'En Proceso'
c.execute("""
    UPDATE auditorias
    SET estado = 'En Proceso', fecha_fin = '2024-03-31'
    WHERE nombre_auditoria = 'Auditoría de Desempeño Gestión Ambiental CAR';
""")
conn.commit()
print("Estado de auditoría y fecha de fin actualizados.")

# Aumentar el monto afectado de un hallazgo específico
c.execute("""
    UPDATE hallazgos
    SET monto_afectado = 180000000.00
    WHERE id_hallazgo = 1;
""")
conn.commit()
print("Monto afectado del hallazgo 1 actualizado.")

# Para verificar los cambios (usaremos SELECT más adelante, por ahora lo haremos directamente)
print("\nVerificando cambios en auditorias (ID 3):")
result_auditoria = c.execute("SELECT * FROM auditorias WHERE id_auditoria = 3;").fetchone()
print(result_auditoria)

print("\nVerificando cambios en hallazgos (ID 1):")
result_hallazgo = c.execute("SELECT * FROM hallazgos WHERE id_hallazgo = 1;").fetchone()
print(result_hallazgo)
````

### 3.3. `DELETE FROM`: Eliminando Datos

El comando `DELETE FROM` se utiliza para eliminar filas existentes de una tabla. Al igual que con `UPDATE`, es fundamental utilizar la cláusula `WHERE` para especificar qué filas deben ser eliminadas. Si se omite la cláusula `WHERE`, ¡todas las filas de la tabla serán eliminadas!

**Ejemplo: Eliminar un hallazgo**

````python
# Eliminar el hallazgo con id_hallazgo = 2
c.execute("""
    DELETE FROM hallazgos
    WHERE id_hallazgo = 2;
""")
conn.commit()
print("Hallazgo con ID 2 eliminado exitosamente.")

# Verificar la eliminación (usaremos SELECT para confirmar)
print("\nVerificando hallazgos restantes:")
result_hallazgos = c.execute("SELECT * FROM hallazgos;").fetchall()
for row in result_hallazgos:
    print(row)
````

### 3.4. Combinación de DML: `INSERT OR REPLACE` y `UPSERT`

En SQLite, una forma común de manejar situaciones donde se quiere insertar una fila si no existe, o actualizarla si ya existe (conocido como UPSERT), es utilizando `INSERT OR REPLACE`. Esto es útil cuando hay claves primarias o únicas y se quiere evitar duplicados o actualizar directamente el registro existente.

**Ejemplo: Insertar un funcionario o actualizarlo si ya existe**

````python
# Intentar insertar un funcionario que ya existe (con mismo id_funcionario)
c.execute("""
    INSERT OR REPLACE INTO funcionarios (id_funcionario, nombre, cargo, departamento, email)
    VALUES
        (1, 'Ana María Gómez', 'Auditor Principal', 'Financiero', 'ana.gomez@contraloria.gov.co');
""")
conn.commit()
print("Funcionario 1 actualizado (o insertado si no existía).")

# Insertar un nuevo funcionario con un ID diferente
c.execute("""
    INSERT OR REPLACE INTO funcionarios (id_funcionario, nombre, cargo, departamento, email)
    VALUES
        (5, 'Elena Castro', 'Analista de Datos', 'Tecnología', 'elena.castro@contraloria.gov.co');
""")
conn.commit()
print("Funcionario 5 insertado.")

# Verificar los cambios en la tabla 'funcionarios'
print("\nLista de funcionarios actualizada:")
result_funcionarios = c.execute("SELECT * FROM funcionarios;").fetchall()
for row in result_funcionarios:
    print(row)
````

## 4. DQL (Data Query Language): Consultando Datos

El DQL se utiliza para recuperar datos de la base de datos. El comando más importante y frecuentemente utilizado es `SELECT`.

### 4.1. `SELECT`: Recuperando Datos Básicos

El comando `SELECT` se utiliza para especificar qué columnas deseas recuperar de una tabla. Puedes seleccionar todas las columnas usando el asterisco (`*`) o listar columnas específicas.

**Ejemplo: Seleccionar todas las columnas de la tabla `auditorias`**

````python
print("\nTodas las auditorías:")
result = c.execute("SELECT * FROM auditorias;").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_auditorias = pd.read_sql_query("SELECT * FROM auditorias;", conn)
print("\nAuditorías (DataFrame):")
display(df_auditorias)
````

**Ejemplo: Seleccionar columnas específicas de la tabla `funcionarios`**

````python
print("\nNombres y cargos de los funcionarios:")
result = c.execute("SELECT nombre, cargo FROM funcionarios;").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_funcionarios_resumen = pd.read_sql_query("SELECT nombre, cargo FROM funcionarios;", conn)
print("\nNombres y Cargos de Funcionarios (DataFrame):")
display(df_funcionarios_resumen)
````

### 4.2. `SELECT DISTINCT`: Eliminando Duplicados

La cláusula `DISTINCT` se utiliza para devolver solo valores únicos (no duplicados) en la columna o columnas especificadas.

**Ejemplo: Obtener los departamentos únicos de los funcionarios**

````python
print("\nDepartamentos únicos de funcionarios:")
result = c.execute("SELECT DISTINCT departamento FROM funcionarios;").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_departamentos = pd.read_sql_query("SELECT DISTINCT departamento FROM funcionarios;", conn)
print("\nDepartamentos Únicos (DataFrame):")
display(df_departamentos)
````

### 4.3. `WHERE`: Filtrando Datos

La cláusula `WHERE` se utiliza para especificar una condición de filtro para las filas que se van a recuperar. Solo las filas que cumplen la condición se incluyen en el conjunto de resultados.

**Ejemplo: Auditorías en estado 'Finalizada'**

````python
print("\nAuditorías Finalizadas:")
result = c.execute("SELECT * FROM auditorias WHERE estado = 'Finalizada';").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_finalizadas = pd.read_sql_query("SELECT * FROM auditorias WHERE estado = 'Finalizada';", conn)
print("\nAuditorías Finalizadas (DataFrame):")
display(df_finalizadas)
````

**Ejemplo: Hallazgos con monto afectado mayor a 100,000,000**

````python
print("\nHallazgos con monto afectado > 100,000,000:")
result = c.execute("SELECT * FROM hallazgos WHERE monto_afectado > 100000000;").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_hallazgos_grandes = pd.read_sql_query("SELECT * FROM hallazgos WHERE monto_afectado > 100000000;", conn)
print("\nHallazgos con Monto Afectado Alto (DataFrame):")
display(df_hallazgos_grandes)
````

### 4.4. Operadores Lógicos en `WHERE` (`AND`, `OR`, `NOT`)

Puedes combinar múltiples condiciones en la cláusula `WHERE` usando los operadores lógicos `AND`, `OR` y `NOT`.

**Ejemplo: Funcionarios del departamento 'Financiero' y con cargo 'Auditor Senior'**

````python
print("\nFuncionarios del departamento Financiero y Auditor Senior:")
result = c.execute("SELECT * FROM funcionarios WHERE departamento = 'Financiero' AND cargo = 'Auditor Senior';").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_fin_senior = pd.read_sql_query("SELECT * FROM funcionarios WHERE departamento = 'Financiero' AND cargo = 'Auditor Senior';", conn)
print("\nFuncionarios Financieros Senior (DataFrame):")
display(df_fin_senior)
````

**Ejemplo: Hallazgos de severidad 'Alta' o 'Grave'**

````python
print("\nHallazgos de severidad Alta o Grave:")
result = c.execute("SELECT * FROM hallazgos WHERE severidad = 'Alta' OR severidad = 'Grave';").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_severidad_alta_grave = pd.read_sql_query("SELECT * FROM hallazgos WHERE severidad = 'Alta' OR severidad = 'Grave';", conn)
print("\nHallazgos de Severidad Alta o Grave (DataFrame):")
display(df_severidad_alta_grave)
````

### 4.5. `ORDER BY`: Ordenando Resultados

La cláusula `ORDER BY` se utiliza para ordenar el conjunto de resultados de una consulta. Puedes ordenar en orden ascendente (`ASC`, por defecto) o descendente (`DESC`).

**Ejemplo: Auditorías ordenadas por fecha de inicio (ascendente)**

````python
print("\nAuditorías ordenadas por fecha de inicio (ASC):")
result = c.execute("SELECT * FROM auditorias ORDER BY fecha_inicio ASC;").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_auditorias_fecha = pd.read_sql_query("SELECT * FROM auditorias ORDER BY fecha_inicio ASC;", conn)
print("\nAuditorías por Fecha de Inicio (DataFrame):")
display(df_auditorias_fecha)
````

**Ejemplo: Hallazgos ordenados por monto afectado (descendente)**

````python
print("\nHallazgos ordenados por monto afectado (DESC):")
result = c.execute("SELECT * FROM hallazgos ORDER BY monto_afectado DESC;").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_hallazgos_monto_desc = pd.read_sql_query("SELECT * FROM hallazgos ORDER BY monto_afectado DESC;", conn)
print("\nHallazgos por Monto Afectado (Descendente, DataFrame):")
display(df_hallazgos_monto_desc)
````

### 4.6. `LIMIT`: Limitando el Número de Resultados

La cláusula `LIMIT` se utiliza para restringir el número de filas devueltas por una consulta `SELECT`. Es útil para paginación o cuando solo necesitas una muestra de los datos.

**Ejemplo: Obtener los 2 primeros funcionarios**

````python
print("\nLos 2 primeros funcionarios:")
result = c.execute("SELECT * FROM funcionarios LIMIT 2;").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_top_funcionarios = pd.read_sql_query("SELECT * FROM funcionarios LIMIT 2;", conn)
print("\nTop 2 Funcionarios (DataFrame):")
display(df_top_funcionarios)
````

### 4.7. Funciones de Agregación (`COUNT`, `SUM`, `AVG`, `MIN`, `MAX`) y `GROUP BY`

Las funciones de agregación realizan un cálculo sobre un conjunto de filas y devuelven un único valor de resumen. La cláusula `GROUP BY` se utiliza para agrupar filas que tienen los mismos valores en columnas especificadas en grupos más pequeños.

**Ejemplo: Contar el número total de auditorías**

````python
print("\nNúmero total de auditorías:")
result = c.execute("SELECT COUNT(*) FROM auditorias;").fetchone()[0]
print(f"Total de auditorías: {result}")
````

**Ejemplo: Suma total de montos afectados en hallazgos**

````python
print("\nSuma total de montos afectados:")
result = c.execute("SELECT SUM(monto_afectado) FROM hallazgos;").fetchone()[0]
print(f"Suma total de montos afectados: {result:,.2f}")
````

**Ejemplo: Monto promedio afectado por severidad de hallazgo**

````python
print("\nMonto promedio afectado por severidad:")
result = c.execute("SELECT severidad, AVG(monto_afectado) FROM hallazgos GROUP BY severidad;").fetchall()
for row in result:
    print(f"Severidad: {row[0]}, Promedio: {row[1]:,.2f}")

# Usando pandas para una visualización más limpia
df_avg_monto_severidad = pd.read_sql_query("SELECT severidad, AVG(monto_afectado) as monto_promedio FROM hallazgos GROUP BY severidad;", conn)
print("\nMonto Promedio por Severidad (DataFrame):")
display(df_avg_monto_severidad)
````

**Ejemplo: Contar el número de funcionarios por departamento**

````python
print("\nNúmero de funcionarios por departamento:")
result = c.execute("SELECT departamento, COUNT(*) FROM funcionarios GROUP BY departamento;").fetchall()
for row in result:
    print(f"Departamento: {row[0]}, Cantidad: {row[1]}")

# Usando pandas para una visualización más limpia
df_funcionarios_depto = pd.read_sql_query("SELECT departamento, COUNT(*) as cantidad_funcionarios FROM funcionarios GROUP BY departamento;", conn)
print("\nFuncionarios por Departamento (DataFrame):")
display(df_funcionarios_depto)
````

### 4.8. `HAVING`: Filtrando Grupos

La cláusula `HAVING` se utiliza para filtrar grupos generados por la cláusula `GROUP BY`, similar a cómo `WHERE` filtra filas individuales. `HAVING` opera en los resultados de las funciones de agregación.

**Ejemplo: Departamentos con más de un funcionario**

````python
print("\nDepartamentos con más de un funcionario:")
result = c.execute("SELECT departamento, COUNT(*) FROM funcionarios GROUP BY departamento HAVING COUNT(*) > 1;").fetchall()
for row in result:
    print(f"Departamento: {row[0]}, Cantidad: {row[1]}")

# Usando pandas para una visualización más limpia
df_depto_mas_uno = pd.read_sql_query("SELECT departamento, COUNT(*) as cantidad_funcionarios FROM funcionarios GROUP BY departamento HAVING COUNT(*) > 1;", conn)
print("\nDepartamentos con más de un Funcionario (DataFrame):")
display(df_depto_mas_uno)
````

### 4.9. `JOIN`: Combinando Datos de Múltiple Tablas

Las cláusulas `JOIN` se utilizan para combinar filas de dos o más tablas, basándose en una columna relacionada entre ellas. Existen varios tipos de JOIN, siendo los más comunes `INNER JOIN`, `LEFT JOIN` (o `LEFT OUTER JOIN`), `RIGHT JOIN` (o `RIGHT OUTER JOIN`) y `FULL JOIN` (o `FULL OUTER JOIN`). SQLite solo soporta `INNER JOIN`, `LEFT JOIN` y `CROSS JOIN` directamente; `RIGHT` y `FULL JOIN` pueden simularse.

**Ejemplo: `INNER JOIN` - Hallazgos y sus Auditorías Correspondientes**

Un `INNER JOIN` devuelve solo las filas que tienen coincidencias en *ambas* tablas.

````python
print("\nHallazgos con el nombre de su auditoría (INNER JOIN):")
result = c.execute("""
    SELECT
        h.descripcion,
        h.severidad,
        a.nombre_auditoria,
        a.estado
    FROM
        hallazgos h
    INNER JOIN
        auditorias a ON h.id_auditoria = a.id_auditoria;
""").fetchall()
for row in result:
    print(row)


# Usando pandas para una visualización más limpia
df_hallazgos_auditorias = pd.read_sql_query("""
    SELECT
        h.descripcion,
        h.severidad,
        a.nombre_auditoria,
        a.estado
    FROM
        hallazgos h
    INNER JOIN
        auditorias a ON h.id_auditoria = a.id_auditoria;
""", conn)
print("\nHallazgos y Auditorías (INNER JOIN, DataFrame):")
display(df_hallazgos_auditorias)
````

**Ejemplo: `LEFT JOIN` - Todas las Auditorías y sus Hallazgos (si existen)**

Un `LEFT JOIN` (o `LEFT OUTER JOIN`) devuelve todas las filas de la tabla de la izquierda, y las filas coincidentes de la tabla de la derecha. Si no hay coincidencia en la tabla de la derecha, los resultados de esa tabla serán `NULL`.

````python
print("\nTodas las auditorías y sus hallazgos (LEFT JOIN):")
result = c.execute("""
    SELECT
        a.nombre_auditoria,
        a.estado,
        h.descripcion,
        h.severidad
    FROM
        auditorias a
    LEFT JOIN
        hallazgos h ON a.id_auditoria = h.id_auditoria;
""").fetchall()
for row in result:
    print(row)


# Usando pandas para una visualización más limpia
df_auditorias_left_hallazgos = pd.read_sql_query("""
    SELECT
        a.nombre_auditoria,
        a.estado,
        h.descripcion,
        h.severidad
    FROM
        auditorias a
    LEFT JOIN
        hallazgos h ON a.id_auditoria = h.id_auditoria;
""", conn)
print("\nAuditorías y Hallazgos (LEFT JOIN, DataFrame):")
display(df_auditorias_left_hallazgos)
````

**Ejemplo: `CROSS JOIN` - Combinación cartesiana (todos contra todos)**

Un `CROSS JOIN` devuelve el producto cartesiano de las filas de las tablas unidas. Es decir, combina cada fila de la primera tabla con cada fila de la segunda tabla. Es poco común y genera resultados muy grandes.

````python
print("\nEjemplo de CROSS JOIN (limitado a 5 filas para no saturar):")
result = c.execute("""
    SELECT
        f.nombre, a.nombre_auditoria
    FROM
        funcionarios f
    CROSS JOIN
        auditorias a
    LIMIT 5;
""").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_cross_join_example = pd.read_sql_query("""
    SELECT
        f.nombre, a.nombre_auditoria
    FROM
        funcionarios f
    CROSS JOIN
        auditorias a
    LIMIT 5;
""", conn)
print("\nCROSS JOIN (Ejemplo limitado, DataFrame):")
display(df_cross_join_example)
````

### 4.10. Subconsultas (Subqueries): Consultas Anidadas

Una subconsulta (o consulta anidada) es una consulta `SELECT` dentro de otra consulta SQL. Se pueden usar en varias cláusulas, como `WHERE`, `FROM`, y `SELECT`.

**Ejemplo: Auditorías con hallazgos de severidad 'Grave'**

Aquí, la subconsulta selecciona los `id_auditoria` de los hallazgos graves, y la consulta externa usa esos IDs para mostrar las auditorías correspondientes.

````python
print("\nAuditorías con al menos un hallazgo de severidad 'Grave' (usando subconsulta):")
result = c.execute("""
    SELECT
        nombre_auditoria,
        estado
    FROM
        auditorias
    WHERE
        id_auditoria IN (SELECT id_auditoria FROM hallazgos WHERE severidad = 'Grave');
""").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_auditorias_graves = pd.read_sql_query("""
    SELECT
        nombre_auditoria,
        estado
    FROM
        auditorias
    WHERE
        id_auditoria IN (SELECT id_auditoria FROM hallazgos WHERE severidad = 'Grave');
""", conn)
print("\nAuditorías con Hallazgos Graves (DataFrame):")
display(df_auditorias_graves)
````

**Ejemplo: Funcionarios que no son 'Auditor Principal'**

````python
print("\nFuncionarios que no son 'Auditor Principal' (usando subconsulta con NOT IN):")
result = c.execute("""
    SELECT
        nombre, cargo
    FROM
        funcionarios
    WHERE
        cargo NOT IN ('Auditor Principal');
""").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_no_auditor_principal = pd.read_sql_query("""
    SELECT
        nombre, cargo
    FROM
        funcionarios
    WHERE
        cargo NOT IN ('Auditor Principal');
""", conn)
print("\nFuncionarios NO Auditores Principales (DataFrame):")
display(df_no_auditor_principal)
````

### 4.11. `UNION`, `INTERSECT`, `EXCEPT`: Combinando Resultados de Consultas

Estas operaciones combinan los resultados de dos o más sentencias `SELECT`.

**Ejemplo: `UNION` - Combinar nombres de auditorías y cargos de funcionarios (demostrativo)**

La cláusula `UNION` combina los conjuntos de resultados de dos o más sentencias `SELECT`. Cada sentencia `SELECT` dentro de `UNION` debe tener el mismo número de columnas, las columnas también deben tener tipos de datos similares y estar en el mismo orden. `UNION` elimina filas duplicadas.

````python
print("\nNombres de auditorías y cargos únicos de funcionarios (UNION):")
result = c.execute("""
    SELECT nombre_auditoria FROM auditorias
    UNION
    SELECT cargo FROM funcionarios;
""").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_union_example = pd.read_sql_query("""
    SELECT nombre_auditoria FROM auditorias
    UNION
    SELECT cargo FROM funcionarios;
""", conn)
print("\nCombinación de Nombres/Cargos (UNION, DataFrame):")
display(df_union_example)
````

**Ejemplo: `INTERSECT` - (Conceptual, SQLite lo soporta directamente)**

La cláusula `INTERSECT` devuelve solo las filas que son comunes a ambos conjuntos de resultados de las sentencias `SELECT`. SQLite tiene un operador `INTERSECT`.

````python
print("\nCargos que son también nombres de auditoría (INTERSECT - si existieran coincidencias lógicas):")
# Demostración conceptual, buscando un cargo que coincida con un estado de auditoría
result = c.execute("""
    SELECT estado FROM auditorias
    INTERSECT
    SELECT cargo FROM funcionarios WHERE cargo = 'Finalizada';
""").fetchall()

if result:
    for row in result:
        print(row)
else:
    print("No se encontraron coincidencias lógicas con INTERSECT en este ejemplo.")

# Usando pandas para una visualización más limpia
df_intersect_example = pd.read_sql_query("""
    SELECT estado FROM auditorias
    INTERSECT
    SELECT cargo FROM funcionarios WHERE cargo = 'Finalizada';
""", conn)
if not df_intersect_example.empty:
    print("\nINTERSECT (DataFrame):")
    display(df_intersect_example)
else:
    print("\nNo se encontraron coincidencias lógicas con INTERSECT en este ejemplo (DataFrame).")
````

**Ejemplo: `EXCEPT` - Auditorías que no tienen hallazgos**

La cláusula `EXCEPT` devuelve solo las filas que son exclusivas del primer conjunto de resultados de las sentencias `SELECT` y no están presentes en el segundo. SQLite tiene un operador `EXCEPT`.

````python
print("\nAuditorías que no tienen ningún hallazgo registrado (EXCEPT):")
result = c.execute("""
    SELECT id_auditoria, nombre_auditoria
    FROM auditorias
    EXCEPT
    SELECT a.id_auditoria, a.nombre_auditoria
    FROM auditorias a JOIN hallazgos h ON a.id_auditoria = h.id_auditoria;
""").fetchall()
for row in result:
    print(row)

# Usando pandas para una visualización más limpia
df_except_example = pd.read_sql_query("""
    SELECT id_auditoria, nombre_auditoria
    FROM auditorias
    EXCEPT
    SELECT a.id_auditoria, a.nombre_auditoria
    FROM auditorias a JOIN hallazgos h ON a.id_auditoria = h.id_auditoria;
""", conn)
print("\nAuditorías sin Hallazgos (EXCEPT, DataFrame):")
display(df_except_example)
````

## 5. DCL (Data Control Language): Gestionando Permisos

El DCL se utiliza para controlar los permisos y el acceso a los datos de la base de datos. Los comandos principales son `GRANT` y `REVOKE`.

**Nota Importante sobre SQLite:**
SQLite es un motor de base de datos embebido y sin servidor, lo que significa que carece de un sistema robusto de usuarios y permisos como los que se encuentran en bases de datos cliente-servidor (como PostgreSQL, MySQL, SQL Server u Oracle). Por lo tanto, los comandos `GRANT` y `REVOKE` no son implementados directamente en SQLite. La gestión de permisos en SQLite se realiza a nivel del sistema de archivos, controlando el acceso al propio archivo de la base de datos.

### 5.1. `GRANT`: Otorgar Permisos (Concepto)

En bases de datos relacionales tradicionales, el comando `GRANT` se usaría para otorgar privilegios a un usuario o rol específico sobre objetos de la base de datos (tablas, vistas, funciones, etc.).

**Sintaxis conceptual (No ejecutable en SQLite directamente):**

```sql
GRANT SELECT, INSERT ON tabla_ejemplo TO usuario_ejemplo;
GRANT ALL PRIVILEGES ON otra_tabla TO otro_usuario;
```

Esto permitiría a `usuario_ejemplo` realizar operaciones `SELECT` e `INSERT` en `tabla_ejemplo`.

### 5.2. `REVOKE`: Revocar Permisos (Concepto)

El comando `REVOKE` se utiliza para retirar privilegios que previamente fueron otorgados a un usuario o rol.

**Sintaxis conceptual (No ejecutable en SQLite directamente):**

```sql
REVOKE INSERT ON tabla_ejemplo FROM usuario_ejemplo;
REVOKE ALL PRIVILEGES ON otra_tabla FROM otro_usuario;
```

Esto eliminaría el permiso de `INSERT` de `usuario_ejemplo` sobre `tabla_ejemplo`.

A pesar de las limitaciones de SQLite en DCL, es fundamental entender que en sistemas de bases de datos de producción, la gestión de permisos es una parte crítica de la seguridad y la integridad de los datos.

## 6. TCL (Transaction Control Language): Gestionando Transacciones

El TCL se utiliza para gestionar transacciones en la base de datos. Una transacción es una secuencia de operaciones que se realizan como una única unidad lógica de trabajo. Todas las operaciones en una transacción deben completarse correctamente, o ninguna de ellas debe hacerlo. Los comandos principales son `COMMIT`, `ROLLBACK` y `SAVEPOINT`.

En `sqlite3` de Python, las transacciones se manejan por defecto con `autocommit` deshabilitado (a menos que se configure lo contrario). Esto significa que cada `c.execute()` que modifica la base de datos (INSERT, UPDATE, DELETE) forma parte de una transacción que debe ser confirmada explícitamente con `conn.commit()` o deshecha con `conn.rollback()`.

### 6.1. `COMMIT`: Confirmar Transacción

El comando `COMMIT` guarda permanentemente todos los cambios realizados en la transacción en la base de datos.

**Ejemplo: Insertar un nuevo funcionario y confirmar**

````python
print("\n--- Demostración de COMMIT ---")
print("Antes de la inserción (contar funcionarios):")
count_before = c.execute("SELECT COUNT(*) FROM funcionarios;").fetchone()[0]
print(f"Total de funcionarios: {count_before}")

c.execute("""
    INSERT INTO funcionarios (nombre, cargo, departamento, email)
    VALUES
        ('Pedro García', 'Analista de Proyectos', 'Auditoría Interna', 'pedro.garcia@contraloria.gov.co');
""")
print("Funcionario 'Pedro García' insertado (aún no confirmado).")

# El cambio no es visible para otras conexiones (o consultas sin autocommit)
# pero sí dentro de la misma transacción.

conn.commit()
print("Cambios confirmados con COMMIT.")

print("Después de la confirmación (contar funcionarios):")
count_after = c.execute("SELECT COUNT(*) FROM funcionarios;").fetchone()[0]
print(f"Total de funcionarios: {count_after}")

df_funcionarios_commit = pd.read_sql_query("SELECT * FROM funcionarios WHERE nombre = 'Pedro García';", conn)
print("\nVerificando funcionario 'Pedro García' (DataFrame):")
display(df_funcionarios_commit)
````

### 6.2. `ROLLBACK`: Deshacer Transacción

El comando `ROLLBACK` deshace todos los cambios realizados en la transacción desde el último `COMMIT` o el inicio de la transacción, restaurando la base de datos a su estado anterior.

**Ejemplo: Insertar un nuevo hallazgo y luego deshacerlo**

````python
print("\n--- Demostración de ROLLBACK ---")
print("Antes de la inserción (contar hallazgos):")
count_before_hallazgos = c.execute("SELECT COUNT(*) FROM hallazgos;").fetchone()[0]
print(f"Total de hallazgos: {count_before_hallazgos}")

c.execute("""
    INSERT INTO hallazgos (id_auditoria, descripcion, severidad, monto_afectado, fecha_identificacion)
    VALUES
        (1, 'Hallazgo experimental que será deshecho', 'Baja', 1000000.00, '2024-01-01');
""")
print("Hallazgo experimental insertado (aún no confirmado).")

print("Después de la inserción, antes del rollback (contar hallazgos - temporal):")
count_after_insert = c.execute("SELECT COUNT(*) FROM hallazgos;").fetchone()[0]
print(f"Total de hallazgos: {count_after_insert}")

conn.rollback()
print("Cambios deshechos con ROLLBACK.")

print("Después del rollback (contar hallazgos):")
count_after_rollback = c.execute("SELECT COUNT(*) FROM hallazgos;").fetchone()[0]
print(f"Total de hallazgos: {count_after_rollback}")

df_hallazgos_rollback = pd.read_sql_query("SELECT * FROM hallazgos WHERE descripcion = 'Hallazgo experimental que será deshecho';", conn)
print("\nVerificando hallazgo experimental (DataFrame - debería estar vacío):")
display(df_hallazgos_rollback)
````

### 6.3. `SAVEPOINT`: Puntos de Guardado (Concepto y uso limitado en `sqlite3`)

La cláusula `SAVEPOINT` permite establecer puntos dentro de una transacción para poder deshacer solo una parte de ella (rollback a un savepoint específico) en lugar de toda la transacción. Esto es útil en transacciones complejas donde se desea flexibilidad en el manejo de errores.

**Sintaxis conceptual (SQLite lo soporta, pero su uso directo en `sqlite3` de Python es menos común, ya que generalmente se gestionan errores de forma más global por transacción o se usan excepciones):**

```sql
BEGIN TRANSACTION;
INSERT INTO ...;
SAVEPOINT my_savepoint;
UPDATE ...;
ROLLBACK TO my_savepoint;
COMMIT;
```

En `sqlite3` de Python, si bien el cursor soporta `execute('SAVEPOINT my_point')`, el enfoque más común es manejar las transacciones en bloque con `conn.commit()` y `conn.rollback()` para toda la unidad de trabajo.

## 7. SQLAlchemy (Opcional: ORM para Python)

SQLAlchemy es un popular kit de herramientas SQL y un Mapeador Objeto-Relacional (ORM) para Python que proporciona una forma de interactuar con bases de datos utilizando objetos Python en lugar de escribir consultas SQL directamente. Esto puede hacer que el código sea más legible, mantenible y menos propenso a errores de inyección SQL. Aunque en este notebook hemos trabajado directamente con `sqlite3` y SQL puro, es importante conocer herramientas como SQLAlchemy para proyectos más grandes y complejos.

**¿Por qué usar SQLAlchemy?**

*   **Abstracción**: Permite trabajar con la base de datos a un nivel más alto, utilizando objetos y clases Python. No necesitas recordar la sintaxis exacta de SQL para cada operación.
*   **Portabilidad**: El mismo código ORM puede funcionar con diferentes bases de datos (SQLite, PostgreSQL, MySQL, SQL Server, etc.) con cambios mínimos, ya que SQLAlchemy abstrae las particularidades de cada dialecto SQL.
*   **Seguridad**: Ayuda a prevenir ataques de inyección SQL al parametrizar automáticamente las consultas.
*   **Productividad**: Reduce la cantidad de código repetitivo (boilerplate) y acelera el desarrollo.
*   **Manejo de Transacciones**: Facilita la gestión de transacciones de manera más idiomática en Python.

A continuación, veremos cómo podríamos haber realizado algunas de las operaciones DDL, DML y DQL que ya cubrimos, pero utilizando SQLAlchemy.

### 7.2. Conexión y Definición de Modelos (DDL con ORM)

Primero, importamos las clases necesarias de SQLAlchemy y establecemos una conexión a nuestra base de datos en memoria. Luego, definimos las tablas como clases Python, que heredan de una `DeclarativeBase`.

````python
from sqlalchemy import create_engine, Column, Integer, String, Float, Date, ForeignKey
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker, relationship
from datetime import date

# Crear un nuevo motor para SQLAlchemy
# Usamos ':memory:' para una base de datos en memoria, similar a la anterior
engine_sa = create_engine('sqlite:///:memory:')

# Base declarativa para nuestros modelos
Base = declarative_base()

# Definir el modelo para la tabla 'auditorias'
class AuditoriaSA(Base):
    __tablename__ = 'auditorias'
    id_auditoria = Column(Integer, primary_key=True, autoincrement=True)
    nombre_auditoria = Column(String, nullable=False)
    fecha_inicio = Column(Date, nullable=False)
    fecha_fin = Column(Date)
    estado = Column(String, nullable=False)

    # Definir la relación con 'HallazgoSA'
    hallazgos = relationship("HallazgoSA", back_populates="auditoria")

    def __repr__(self):
        return f"<AuditoriaSA(id={self.id_auditoria}, nombre='{self.nombre_auditoria}', estado='{self.estado}')>"

# Definir el modelo para la tabla 'funcionarios'
class FuncionarioSA(Base):
    __tablename__ = 'funcionarios'
    id_funcionario = Column(Integer, primary_key=True, autoincrement=True)
    nombre = Column(String, nullable=False)
    cargo = Column(String, nullable=False)
    departamento = Column(String, nullable=False)
    email = Column(String) # Columna añadida previamente con ALTER TABLE

    def __repr__(self):
        return f"<FuncionarioSA(id={self.id_funcionario}, nombre='{self.nombre}', cargo='{self.cargo}')>"

# Definir el modelo para la tabla 'hallazgos'
class HallazgoSA(Base):
    __tablename__ = 'hallazgos'
    id_hallazgo = Column(Integer, primary_key=True, autoincrement=True)
    id_auditoria = Column(Integer, ForeignKey('auditorias.id_auditoria'), nullable=False)
    descripcion = Column(String, nullable=False)
    severidad = Column(String, nullable=False)
    monto_afectado = Column(Float)
    fecha_identificacion = Column(Date, nullable=False)

    # Definir la relación con 'AuditoriaSA'
    auditoria = relationship("AuditoriaSA", back_populates="hallazgos")

    def __repr__(self):
        return f"<HallazgoSA(id={self.id_hallazgo}, auditoria_id={self.id_auditoria}, severidad='{self.severidad}')>"

print("Modelos de SQLAlchemy definidos para AuditoriaSA, FuncionarioSA y HallazgoSA.")
````

Una vez definidos los modelos, podemos crear las tablas en la base de datos utilizando `Base.metadata.create_all(engine_sa)`.

````python
# Crear todas las tablas definidas en la Base
Base.metadata.create_all(engine_sa)
print("Tablas creadas en la base de datos SQLAlchemy.")
````

### 7.3. Insertando Datos (DML con ORM)

Para interactuar con la base de datos, usamos una *sesión*. La sesión es la interfaz principal para el almacenamiento de objetos. Permite persistir objetos nuevos, cargar objetos existentes y realizar consultas.

````python
# Crear una sesión para interactuar con la base de datos
SessionSA = sessionmaker(bind=engine_sa)
session_sa = SessionSA()

# Insertar datos en la tabla 'auditorias'
a1 = AuditoriaSA(nombre_auditoria='Auditoría Financiera 2023 MinHacienda SA', fecha_inicio=date(2023, 1, 15), fecha_fin=date(2023, 6, 30), estado='Finalizada')
a2 = AuditoriaSA(nombre_auditoria='Auditoría de Cumplimiento Contratación Pública IDU SA', fecha_inicio=date(2023, 3, 1), fecha_fin=date(2023, 9, 15), estado='En Proceso')
a3 = AuditoriaSA(nombre_auditoria='Auditoría de Desempeño Gestión Ambiental CAR SA', fecha_inicio=date(2023, 5, 10), estado='Planeada')

session_sa.add_all([a1, a2, a3])
session_sa.commit()
print("Datos insertados en la tabla 'auditorias' (SQLAlchemy).")

# Insertar datos en la tabla 'funcionarios'
f1 = FuncionarioSA(nombre='Ana María Gómez SA', cargo='Auditor Principal', departamento='Financiero', email='ana.gomez.sa@contraloria.gov.co')
f2 = FuncionarioSA(nombre='Carlos Pérez Rojas SA', cargo='Jefe de Equipo', departamento='Contratación', email='carlos.perez.sa@contraloria.gov.co')

session_sa.add_all([f1, f2])
session_sa.commit()
print("Datos insertados en la tabla 'funcionarios' (SQLAlchemy).")

# Insertar datos en la tabla 'hallazgos' (relacionados con auditorías)
h1 = HallazgoSA(auditoria=a1, descripcion='Deficiencias en conciliación SA', severidad='Alta', monto_afectado=200000000.00, fecha_identificacion=date(2023, 4, 25))
h2 = HallazgoSA(auditoria=a2, descripcion='Contrato sin publicidad SA', severidad='Grave', monto_afectado=6000000000.00, fecha_identificacion=date(2023, 8, 5))

session_sa.add_all([h1, h2])
session_sa.commit()
print("Datos insertados en la tabla 'hallazgos' (SQLAlchemy).")
````

### 7.4. Consultando Datos (DQL con ORM)

Las consultas se realizan utilizando el objeto `session` y los modelos definidos. SQLAlchemy traduce las operaciones de objetos Python a SQL.

````python
# Consultar todas las auditorías
print("\nTodas las auditorías (SQLAlchemy):")
for auditoria in session_sa.query(AuditoriaSA).all():
    print(auditoria)

# Consultar funcionarios con filtro
print("\nFuncionarios del departamento 'Financiero' (SQLAlchemy):")
for funcionario in session_sa.query(FuncionarioSA).filter_by(departamento='Financiero').all():
    print(funcionario)

# Consultar hallazgos con JOIN (accediendo a la relación)
print("\nHallazgos y sus auditorías (SQLAlchemy JOIN):")
for hallazgo in session_sa.query(HallazgoSA).join(AuditoriaSA).filter(AuditoriaSA.estado == 'Finalizada').all():
    print(f"Hallazgo: {hallazgo.descripcion}, Auditoría: {hallazgo.auditoria.nombre_auditoria}")
````

### 7.5. Actualizando Datos (DML con ORM)

Para actualizar, primero se consulta el objeto, se modifican sus atributos y luego se confirma la sesión.

````python
# Actualizar el estado de una auditoría
auditoria_a_actualizar = session_sa.query(AuditoriaSA).filter_by(nombre_auditoria='Auditoría de Desempeño Gestión Ambiental CAR SA').first()
if auditoria_a_actualizar:
    auditoria_a_actualizar.estado = 'En Proceso'
    session_sa.commit()
    print(f"\nEstado de auditoría '{auditoria_a_actualizar.nombre_auditoria}' actualizado a '{auditoria_a_actualizar.estado}'.")

# Verificar el cambio
print("Auditoría actualizada:", session_sa.query(AuditoriaSA).filter_by(nombre_auditoria='Auditoría de Desempeño Gestión Ambiental CAR SA').first())
````

### 7.6. Eliminando Datos (DML con ORM)

Similar a la actualización, se consulta el objeto a eliminar y se utiliza `session.delete()` antes de confirmar.

````python
# Eliminar un funcionario
funcionario_a_eliminar = session_sa.query(FuncionarioSA).filter_by(nombre='Carlos Pérez Rojas SA').first()
if funcionario_a_eliminar:
    session_sa.delete(funcionario_a_eliminar)
    session_sa.commit()
    print(f"\nFuncionario '{funcionario_a_eliminar.nombre}' eliminado.")

# Verificar la eliminación
print("Funcionarios restantes:", session_sa.query(FuncionarioSA).all())

# Cerrar la sesión de SQLAlchemy
session_sa.close()
print("\nSesión de SQLAlchemy cerrada.")
````
